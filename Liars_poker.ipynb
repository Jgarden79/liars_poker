```python
def ports():
    s = pd.read_pickle('master.pkl')
    names = s.index.drop_duplicates().to_list()

    for n in names:
        df = s[s.index == n]
        df = df.filter(['Stocks', 'Investment', 'Cost per Share', 'DSPLY_NAME', 'GICS Sector Name', 'Close Price', 'Todays Value','Total Return'])
        df['Price Return'] = (df['Close Price']/df['Cost per Share'])-1
    
        df_tots = pd.DataFrame({'Name': n, 'Total Invested':df['Investment'].sum(), "Today's Value": df['Todays Value'].sum()}, index = [0])
        df_tots['Total Return'] =  ((df_tots["Today's Value"])/df_tots['Total Invested'])-1
    
        fig = make_subplots(
            rows = 2, cols = 1,
            vertical_spacing=0.02,
            row_heights = [0.75, 0.25],
            specs=[[{"type": "table"}],
                   [{"type": "table"}]])
    
    
        width = [50, 50, 50, 100, 120, 50, 50, 50]   
                                                                                               
        rowEvenColor = 'lightgrey'
        rowOddColor = 'white'
        fig.append_trace(go.Table(
            columnwidth=width,
            header = dict(values = list(df.columns),
                          fill_color = '#003C5A',
                          font = dict(family= 'pt sans narrow', size = 20),
                          font_color = 'white',
                          align = 'center'),
            cells = dict(values=[df['Stocks'],df['Investment'], df['Cost per Share'],df['DSPLY_NAME'], df['GICS Sector Name'], df['Close Price'],df['Todays Value'], df['Total Return'], df['Price Return']],
                         fill_color = [[rowOddColor, rowEvenColor]*50],
                         font = dict(family= 'pt sans narrow', size = 18),
                         height = 30,
                         format = [[None], ['$,.2f'], ['$,.2f'],[None],[None],['$,.2f'],['$,.2f'],['.2%'], ['.2%']],
                         align = ['left'])), row=1, col = 1)

                                                                                               
        fig.append_trace(go.Table(go.Table(
            columnwidth=width,
            header = dict(values = list(df_tots.columns),
                          fill_color = '#003C5A',
                          font = dict(family= 'pt sans narrow', size = 20),
                          font_color = 'white',
                          align = 'center'),
            cells = dict(values=[df_tots['Name'],df_tots['Total Invested'], df_tots["Today's Value"],df_tots['Total Return']],
                         fill_color = [[rowOddColor, rowEvenColor]*50],
                         font = dict(family= 'pt sans narrow', size = 18),
                         height = 30,
                         format = [[None], ['$,.2f'], ['$,.2f'],['.2%']],
                         align = ['center']))),row=2, col = 1)
        fig.update_layout(
            margin=dict(l=10, r=10, t=40, b=10),
            height = 562,
            width = 1000,
            title="{} Portfolio".format(n))                                                                                            
                                                                                             
        #py.iplot(fig_6, filename='fig_6')
        pio.write_html(fig, file='{}_liars.html'.format(n), auto_open=True)
    
    
    
```
